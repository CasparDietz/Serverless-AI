```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
```

```python
# Funtion that converts a list of strings to a list of floats
def convertToFloat(list):
    return [float(i) for i in list]
```

```python
# Fucntion that takes the list of lists and plots the data
def plotData(MLTime, TotalTime, RoundTrip, FrameRate):
    numberThreads = len(MLTime)
    print(numberThreads)
    
    fig, axs = plt.subplots(3, sharex=True, sharey=True)
    
 
    fig.suptitle('Elapsed Time per Frame: %i Thread(s); %s Frame Rate' % (numberThreads, FrameRate))
    for i in range(numberThreads):
        axs[0].plot(RoundTrip[i], label='Round Trip Time', alpha = 1-(i/numberThreads), color='blue')
        axs[0].set_title('Round Trip Time')

        axs[1].plot(TotalTime[i], label='Total Time', alpha = 1-(i/numberThreads), color='red')
        axs[1].set_title('Total Time')

        axs[2].plot(MLTime[i], label='ML Time', alpha = 1-(i/numberThreads), color='green')
        axs[2].set_title('ML Time')

    axs[2].set_xlabel('Number of Frame')
    for ax in axs.flat:
        ax.set(ylabel='Time (s)')
        #ax.grid()
    
```

```python
# Fucntion that takes the list of lists and plots the data
def plotAllData(MLTimeFullFR, MLTimeHalfFR, 
                      TotalTimeFullFR, TotalTimeHalfFR, 
                      RoundTripFullFR, RoundTripHalfFR):
    numberThreads = len(MLTimeFullFR)
    print(numberThreads)
    
    fig, axs = plt.subplots(3, 2, sharex=True, sharey=True, figsize=(12,5))
    fig.subplots_adjust(hspace=0.4, wspace=0.1)
    
    fig.suptitle('Elapsed Time per Frame: %i Thread(s)' % (numberThreads))
    for i in range(numberThreads):
        axs[0][0].plot(RoundTripFullFR[i], label='Round Trip Time', alpha = 1-(i/numberThreads), color='blue')
        axs[0][0].set_title('Round Trip Time')
        axs[0][1].plot(RoundTripHalfFR[i], label='Round Trip Time', alpha = 1-(i/numberThreads), color='blue')
        axs[0][1].set_title('Round Trip Time')

        axs[1][0].plot(TotalTimeFullFR[i], label='Total Time', alpha = 1-(i/numberThreads), color='red')
        axs[1][0].set_title('Total Time')
        axs[1][1].plot(TotalTimeHalfFR[i], label='Total Time', alpha = 1-(i/numberThreads), color='red')
        axs[1][1].set_title('Total Time')

        axs[2][0].plot(MLTimeFullFR[i], label='ML Time', alpha = 1-(i/numberThreads), color='green')
        axs[2][0].set_title('ML Time')
        axs[2][1].plot(MLTimeHalfFR[i], label='ML Time', alpha = 1-(i/numberThreads), color='green')
        axs[2][1].set_title('ML Time')



 
```
